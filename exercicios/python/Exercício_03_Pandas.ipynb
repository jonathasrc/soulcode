1 - Crie um DataFrame com as colunas nome, idade, altura, peso e sexo usando listas.

```python
# criando numpy array por motivo de dados homogêneos
names = np.array(['João', 'Maria', 'José', 'Ana', 'Pedro'])
ages = np.array([25, 30, 35, 40, 45])
heights = np.array([1.75, 1.65, 1.80, 1.60, 1.70])
weights = np.array([75, 65, 80, 98, 70])
genders = np.array(['M', 'F', 'M', 'F', 'M'])

# criando um dataframe pandas atraves de lista
data = {
    'name': names,
    'age': ages,
    'height': heights,
    'weight': weights,
    'gender': genders
}
df = pd.DataFrame(data)
```
