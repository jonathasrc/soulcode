## Exploração dos dados

**pd.value_counts()**: Conta as frequencias dos dados

**outliers**: os valores que são considerados alto e baixo dos restantes dos dados.

- Vamos falar que 75% is Q3 e value do Q1 é 25%;
- Outliers são menores que  Q1 - 1.5(Q3-Q1) e maiores que Q3 + 1.5(Q3-Q1).(Q3-Q1) = IQR. Vamos utilizar o **describe()** metodo.

- **describe()** inclui os metodos seguintes:
  - **count**: Numeros de entradas
  - **mean**: Média das entradas
  - **std**: Desvio padrão
  - **min**: Menor valor da entrada
  - **25%**: Primeiro quartil
  - **50%**: Mediana e segundo quartil
  - **75%**: Terceiro quartil
  - **max**: Maior entrada

```python
def clean_column_names(df):
    df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

    for col in df.columns:
        df.rename(columns={col:unidecode(col)}, inplace=True)

    return df

clean_column_names(df)
df.info()
```

### 1. Taxa de Rotavidade

```python
# Taxa de Rotatividade ->
#taxa_rotatividade = (numeros_uncionarios_nao_ativos)/ numeros_funcionarios) * 100

df_inactive_employees = df['demissao'].value_counts()
display(df_inactive_employees)
rate_turnover  = df_inactive_employees['Sim'] / len(df) * 100


df_inactive_employees.to_csv('inactive_employees.csv')
print('-'*80)
print(f'Taxa de rotatividade: {rate_turnover:.2f}')
```

### 2. Rotavidade por experiencia de funcionario

```python
df_transform = df[['demissao','total_de_anos_trabalhados_na_empresa']]

# display(df_transform['total_de_anos_trabalhados_na_empresa'].describe())
bins = [0, 3, 5, 9, 40]
labels = ["Novatos", "Iniciante", "Experiente", "Senior"]

df_transform['categoria'] = pd.cut(df_transform['total_de_anos_trabalhados_na_empresa'], bins=bins, labels=labels, right=False)
df_transform_filtered_inactive_employess = df_transform[df['demissao'] == 'Sim']
total_inactive_employess = len(df_transform_filtered_inactive_employess)

print(f'Total de funcionarios desligados: {total_inactive_employess}')

# display(df_transform_filtered_inactive_employess.value_counts('categoria'))

total_employess_by_category = df_transform_filtered_inactive_employess['categoria']
total_employess_by_category = total_employess_by_category.value_counts()
df_result = total_employess_by_category.rename('total funcionarios desligados por experiencia')

df_result.to_csv('df_rotatividade_por_experiencia.csv')
df_result
```

### 3. Quantidade de Rotatividade por genero



```python
df_turnover_gender = df.loc[:,('genero','demissao')]
df_filtered_gender_count = df_turnover_gender.loc[:, 'genero'].value_counts()
# display(df_filtered_gender_count)


df_result = pd.crosstab(df_turnover_gender['genero'],df_turnover_gender['demissao'])
df_result.to_csv('df_quantidade_rotatividade_por_genero.csv')
display(df_result)




```

### 4. Nivel de maturidade dos coloaboradores da empresa

```python
'''
Nivel de maturidade dos coloaboradores da empresa
https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy


'''
df_transform = df.loc[:,('total_de_anos_trabalhados_na_empresa','demissao')]
bins = [0, 3, 5, 9, 40]
labels = ["Novatos", "Iniciante", "Experiente", "Senior"]

df_transform['categoria'] = pd.cut(df_transform['total_de_anos_trabalhados_na_empresa'],
                                   bins=bins,
                                   labels=labels,
                                   right=False)

df_transform_filtered_active_employess = df_transform[df['demissao'] == 'Não']
total_active_employess = len(df_transform_filtered_active_employess)

# To normalize value counts in a pandas Series, use the normalize argument.
df_percentage_active_employess = df_transform_filtered_active_employess['categoria'].value_counts(
    normalize=True).mul(100).round(2).rename('%')

print(f'Total de funcionarios ativos: {total_active_employess}')
df_result = pd.concat([df_transform_filtered_active_employess.value_counts('categoria'),
                       df_percentage_active_employess],axis=1)

df_result.to_csv('df_nivel_maturidades dos colabores da empresa.csv')
df_result
```

### 5. Perfil Demográfico dos Funcionários: Distribuição Etária [Adulto, Adulto Senior, Senior]|

```python
df_demographic_age = df.loc[:,('genero', 'idade')]

df_demographic_age['idade'].describe()

labels = ['Adulto', 'Adulto Senior', 'Senior']
bins = [18, 54, 64, 100]

df_demographic_age['faixa_etaria'] = pd.cut(df_demographic_age['idade'],
                                            bins=bins,
                                            labels=labels,
                                            right=False)

df_result = df_demographic_age['faixa_etaria'].value_counts(normalize=True).rename('%').mul(100).round(2)
df_result.to_csv('df_perfil_demografico_dos_funcionarios.csv')
df_result
```

## 15. Quantidade de funcionario categoria por Distância até o Trabalho

```python
df_distance_work = df.loc[df['demissao'] == 'Não' ,('id_funcionario','distancia_de_casa')]
# df_distance_work['distancia_de_casa'].describe()

bins = [ 0, 14, 21, float('inf')] # float('inf') infinito positivo
labels = ['proximo', 'medio', 'distante']

df_distance_work['categoria'] = pd.cut(df_distance_work['distancia_de_casa'],
                                       bins=bins,
                                       labels=labels)
df_result = (df_distance_work['categoria'].value_counts().rename('quantidade'))
df_result.to_csv('df_quantidade_funcionario_categoria_distancia_trabalho.csv')
display(df_result)
```

## 16. Rotatividade X Distância até o Trabalho

```python
df_distance_work_2 = df.loc[df['demissao'] == 'Sim',('id_funcionario','distancia_de_casa','demissao')]
total_inactive_employess = len(df_distance_work_2)

bins = [ 0, 14, 21, float('inf')] # float('inf') infinito positivo
labels = ['proximo', 'medio', 'distante']

df_distance_work_2['categoria'] = pd.cut(df_distance_work_2['distancia_de_casa'],
                                       bins=bins,
                                       labels=labels)

df_distance_work_2 = df_distance_work_2.groupby('categoria').size().rename('quantidade')
df_distance_work_2 = df_distance_work_2.to_frame()
df_distance_work_2['percentual'] = (df_distance_work_2['quantidade']
                                            / total_inactive_employess * 100).round(2)

df_distance_work_2.to_csv('df_rotatividade_x_distancia_trabalho.csv')
df_distance_work_2
```
